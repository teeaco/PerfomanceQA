**Веб-разработка**
- фронт
- бэк
- базы данных
- девопс

Юзер-агенты (обмен информацией) 
- браузеры
- мобильный приложения
- скрипты, боты и другие

DNS-серверы (связывает доменные имена с IP-адресами)
1. Пользователь вводит URL-адрес
2. Браузер проверяет локальный кэш на сооответствие 
3. Отправка запроса на DNS-сервер
4. Поиск IP-адреса

Веб-серверы (основные функции):
- Прием входящих запросов
- Обработка запросов
- Отправка статических файлов
- Управление доступом

Уровни кэширования:
- DNS-кэширование
- Кэширование на уровне CDN
- на стороне клиента
- на сервере


Архитектуры веб-приложений:
1. Монолит
Все компоненты тесно связаны и работают как одно целое
2. Микросервисы
Разбивает на набор небольших сервисов, каждый отвечает за конкретную функцию.

Взаимодействие:
1. Инициация запроса
2. Формирование http-запроса
3. DNS-разрешение
4. Установление соединения
5. Обработка запроса веб-сервером
6. Маршрутизация и бизнес-логика
7. Взаимодействие с базой данных
8. Обработка результата (получение данных и обработка)
9. Формирование HTTP-ответа (статус кода состояния, заголовки, тип возвращ контента, куки, кэш-контроль, тело ответа)
10. Отправка ответа клиенту
11. Обработка ответа браузером
12. Кэширование на стороне клиента
